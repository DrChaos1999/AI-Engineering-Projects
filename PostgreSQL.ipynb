2. What does "relational database" mean?

A realtional database organizes data into tables that can be connected through realtionships.

For example:

`customers
--------------------------
id | name
1  | Rahim
2  | Karima
`

`orders
--------------------------------
id | customer_id | total_amount
1  | 1           | 1500
2  | 1           | 800
3  | 2           | 2200
`

The `Customer_ID` in the `orders` table connects each order to a customer.

`customers.id
      ↑
orders.customer_id`


This realtionship lets us ask questions such as:

`Which orders belong to Rahim?`

`How much money has each customer spent?`

`Which customers have never placed an order?`

# 21. Foreign Keys:

A foreign key creates a relationship between tables.

Creates an order table:

`
</>SQL
CREATE TABLE orders (
    id INTEGER GENERATED ALWAYS AS IDENTITY PRIMARY KEY,
    customer_id INTEGER NOT NULL,
    total_amount NUMERIC(10,2)  NOT NULL,
    status VARCHAR(30) DEFAULT 'pending',
    creadted_at TIMESTAMPTZ DEFAULT CURRENT_TIMESTAMP,

    CONSTANT fk_customer
        FOREIGN KEY (customer_id)
        REFERENCES customers(id)
);
This prevents an order from referring to a customer who does not exist. Insert an order:

`
</>SQL
INSERT INTO orders (customer_id,total_amount)
VALUES (1,1500.00);
`

This would fail if customer `1` did not exist.


# 22. Foreign-key deletion behavior
Suppose a customer has orders. What happens if the customer is deleted?
You can define the behavior.

Prevent deletion:

`
</>SQL
REFERENCES customers(id)
On DELETE RESTRICT
`


Delete realted orders automatically

`
</>SQL
REFERENCES customers(id)
ON DELETE SET NULL
`
Choose these rules carefully because they affect data integrity.


# 23. Joins

Joins combine related rows from multiple tables.

Inner join: Returns only matching records:

`
</>SQL
SELECT
    customers.name,
    orders.total_amount,
    orders.status
FROM customers
JOIN orders
    ON customers.id = orders.customer_id;
`
RESULT:

`
name    | total_amount | status
Rahim   | 1500.00      | pending
Karima  | 2200.00      | completed
`

Left Join:
Returns every customer, even if they have no orders:

`
</>SQL
SELECT
    customers.name,
    orders.total_amount
FROM customers
LEFT JOIN orders
    ON customers.id = orders.customer_id;
`

Join idea

`customer.id = orders.cusmtomer_id` -> Joins are one of the most important relational database concepts.

# 30. Transactions:

A transaction groups operations together.
Suppose money is transferred from one account to another:

`
</>SQL
BEGIN;
UPDATE accounts
SET balance = balance -500
WHERE id = 1;


UPDATE accounts
SET balance = balance + 500
WHERE id = 2;
COMMIT;
`

If something goes wrong:

`
</>SQL
ROLLBACK;
`
A transaction ensures that either:

`Both updates happen`

OR:

`Neighther update happens`

This is known as atomicity.

# 31. ACID properties:

PostgreSQL transactions follow the ACID principles.


Atomicity

All operations succeed together or fail together.

Consistency:

The database moves from one valid state to another.

Isolation:

Concurrent transactions should not incorrectly interfere with each other.

Durability:

Commited data remains saved even after a restart or failure.
These properties are a major PostgreSQL is trusted for production applications.

# 34. Views

A view is saved query.

Create a customer spending view:

`
</>SQL
CREATE VIEW customer_spending AS
SELECT 
    customers.id,
    customers.name,
    COALESCE(SUM(orders.total_amount),0) AS total_spending
FROM customers
LEFT JOIN orders
    ON customers.id = orders.customer_id
GROUP BY customers.id,customers.name;
`
Query the view:

`
</>SQL
SELECT *
FROM customer_spending;
`
A normal view stores the query definition, not a separte copy of the data.
